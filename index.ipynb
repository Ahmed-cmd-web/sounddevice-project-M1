```python
import numpy as np
import sounddevice as sd
import matplotlib.pyplot as plt





t = np.linspace(0 , 3 , 12 * 1024)
thirdOctave={'C3':130.81,'D3':146.83,'E3':164.81,
      'F3':174.61,'G3':196,'A3':220,'B3':246.93,}
fourthOctave = {
    'C4':261.63,'D4':293.66,'E4':329.63,
    'F4':349.23,'G4':392.00,'A4':440.00,
    'B4':493.88
}


f1=list(thirdOctave.values())
f2=list(fourthOctave.values())


ti = [0.4, 0.4, 0.8, 1.0, 1.5, 2.0, 2.6]
Ti = [0.05, 0.7, 0.5, 0.3, 0.4, 0.3, 0.2]
N = 7

final=[]

u=lambda t,shift:np.heaviside(-shift+t,0) 
x=lambda f1,f2:(np.sin(2*np.pi*f1*t)+np.sin(2*np.pi*f2*t)) 

res=0

for j in range(N):
    f1=list(thirdOctave.values())
    f2=list(fourthOctave.values())
    randomF1Index=np.random.randint(0,len(f1))
    randomF2Index=np.random.randint(0,len(f2))
    res+=x(f1[randomF1Index],f2[randomF2Index]) * ((t>=ti[j])&(t<=(ti[j]+Ti[j])))
    

plt.plot(t,res)


sd.play(res, 3*1024)









```
